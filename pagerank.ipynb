```python
import pandas as pd
import numpy as np
```

```python
data =  pd.read_csv('src/assets/cluster2frame.csv')
ct = pd.crosstab(index = data.Source, columns = data.Target)
```

```python
data
```

```python
sset = set(np.unique(data.Source))
len(sset)
```

```python
tset = set(np.unique(data.Target)) 
len(tset)
```

```python
diff = tset.difference(sset)
```

```python
ct_out = ct
```

```python
ct_out = ct_out.drop(diff, axis=1)
ct_out
```

```python
def pagerank(matrix, gamma, n_iter=None):
    print(matrix.shape)
    tele = np.ones((np.alen(matrix), np.alen(matrix[0])))
    tele = np.matrix([t/sum(t) for t in tele])
    rw = []
    for row in matrix:
        if(all(row == 0)):
            rw.append(row)
        else:
            rw.append(row/sum(row))
    rw = np.matrix(rw)
    P = (rw*gamma) + ((1-gamma)*tele)
    if n_iter is not None:
        for i in range(0, n_iter):
            P = P*P
            print(P)
    else:
        P_before = np.zeros(P.shape)
        while not all((P_before-P) == 0):
            P_before = P
            P = P*P
            print(P)
    return P
```

```python
m = np.array([[1, 0, 1], [0, 0, 1], [0,1,0]])
pagerank(m, 0.5, 5)
```

```python
rank = pagerank(ct_out.as_matrix(), 0.85,5)
rank
```

```python
data_out = pd.DataFrame(data=np.squeeze(np.asarray(rank[0])), index=ct_out.columns)
```

```python
data_out.to_csv('src/assets/pagerank68')
```
