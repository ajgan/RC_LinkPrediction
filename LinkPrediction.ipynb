```python
import pandas as pd
import numpy as np
```

```python
def degree(vertex, matrix):
    return np.sum(matrix[vertex])
```

```python
def inter(v1, v2, matrix):
    r1 = matrix[matrix.index[v1]]
    r2 = matrix[matrix.index[v2]]
    return np.multiply(r1,r2)
```

```python
def adamicAdar(v1, v2, matrix):
    som = 0
    dum = inter(v1,v2,matrix)
    for i in range(len(dum)):
        if(dum[i] == 1):
            item = 1/np.log10(degree(i,matrix))
            som += item
    return som
```
